```python
import pandas as pd 
df=pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv');
df
```

```python
plt.figure(figsize=(8,6))
sns.set_style("darkgrid")
sns.set_context("notebook")
plt.scatter(data=df,x="sepal_length",y="petal_length",marker="v",linewidths=2)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x="sepal_length",y="petal_length",hue='species')
```

```python
plt.style.use('default')
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x="sepal_length",y="petal_length",hue='species',style='species')
```

```python
fig_dims=(10,7)
fig,ax=plt.subplots(figsize=fig_dims)
sns.scatterplot(data=df,x="sepal_width",y="sepal_length",hue="species",style='species',size='species',ax=ax,sizes=(40,300),linewidth=2)
plt.legend(loc='lower right',bbox_to_anchor=(1.2,0.5))#shadow=True
plt.show()
```

# strip plot

```python
exercise=pd.read_csv('D:\\python_lab_20004\\test\\exercise.csv')
exercise.head()
```

```python
plt.figure(figsize=(7,7))
sns.stripplot(x="kind",y="pulse",hue="diet",size=9,data=exercise,palette="cool",dodge=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(x="kind",y="pulse",hue="diet",size=9,data=exercise,dodge=True)
plt.show()
```
